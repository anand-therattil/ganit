```python
x= np.linspace(-5, 5, 5)
print(x)
y = np.exp(np.linspace(-5, 5, 5))
print(y)
plt.plot(x, y)
plt.show()
```

```python
w = np.zeros(len(x))
print(w)
```

```python
predicted = x*w + np.random.normal(0, 1, len(x))
print(predicted)
```

```python
loss = np.mean((predicted - y)**2)
print(loss)
```

```python

# GRADIENT DESCENT IMPLEMENTATION
learning_rate = 0.1
max_iterations = 100
```

```python

for iteration in range(max_iterations):
    # Calculate predictions
    predicted = x * w
    
    # Calculate error
    error = predicted - y
    
    # Calculate loss
    loss = np.mean(error**2)
    
    # Calculate gradient
    gradient = 2 * error * x / len(x)
    
    # Update weights
    w = w - learning_rate * gradient
    
    # Print progress
    if iteration % 5 == 0:
        print(f"Iteration {iteration}: Loss = {loss:.6f}")

print(f"Final loss: {loss:.8f}")
print(f"Final weights: {w}")
```
